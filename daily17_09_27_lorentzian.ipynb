```python
%matplotlib inline
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Power spectrum of an exponential decay signal:
http://pages.jh.edu/~signals/spectra/spectra.html

```python
def lorentzian(x, params):
    a,b,c = params
    return a/(b+x**c)
```

```python
x = np.arange(0,400,0.1)
for pm in np.arange(0,10,1)**2:
    y = lorentzian(x,(1.,pm, 2.))
    plt.loglog(x,y)
```
